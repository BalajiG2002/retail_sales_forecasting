```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
import os

def load_and_prepare_data(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Standardize column names
    df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
    
    # Create a mapping of expected column names
    expected_columns = {
        'date': 'date',
        'store_id': 'store_id',
        'store_name': 'store_name',
        'item_id': 'item_id',
        'item_name': 'item_name',
        'ad_spend': 'ad_spend',
        'units_sold': 'units_sold',
        'units': 'units_sold',
        'ordered_revenue_amount': 'revenue',
        'revenue': 'revenue',
        'unit_price': 'unit_price',
        'anarix_id': 'anarix_id'
    }
    
    # Map columns to a standardized format
    df = df.rename(columns={col: name for col, name in expected_columns.items() if col in df.columns})
    
    # Drop columns not in the expected set, but keep all necessary ones
    necessary_columns = ['date', 'store_id', 'ad_spend', 'units_sold', 'revenue', 'unit_price']
    df = df[[col for col in necessary_columns if col in df.columns]]
    
    # Convert 'date' column to datetime and set it as index
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    
    return df

# Load and prepare the dataset
df = load_and_prepare_data('/home/balajig/retail_sales_forecasting/data/synthetic_retail_sales.csv')

# Basic Data Overview
print("Data Head:")
print(df.head())

print("\nData Info:")
print(df.info())

print("\nData Description:")
print(df.describe())

# Check for Missing Values
print("Missing Values:")
print(df.isnull().sum())

# Distribution of Units Sold
plt.figure(figsize=(10, 6))
sns.histplot(df['units_sold'], kde=True)
plt.title('Distribution of Units Sold')
plt.show()

# Time Series Plot of Units Sold
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x=df.index, y='units_sold')
plt.title('Units Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.show()

# Seasonality Check: Monthly
df['month'] = df.index.to_period('M').astype(str)
monthly_sales = df.groupby(['month'])['units_sold'].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_sales, x='month', y='units_sold')
plt.title('Monthly Units Sold')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()

# Feature Engineering
df['year'] = df.index.year
df['quarter'] = df.index.quarter
df['month'] = df.index.month
df['day_of_week'] = df.index.dayofweek

# Lag features
df['lag_1'] = df['units_sold'].shift(1)
df['lag_2'] = df['units_sold'].shift(2)
df['lag_3'] = df['units_sold'].shift(3)

# Rolling statistics
df['rolling_mean_3'] = df['units_sold'].rolling(window=3).mean()
df['rolling_std_3'] = df['units_sold'].rolling(window=3).std()
df.dropna(inplace=True)

# Target and Features
X = df[['ad_spend', 'lag_1', 'lag_2', 'lag_3', 'rolling_mean_3', 'rolling_std_3', 'month', 'day_of_week']]
y = df['units_sold']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
print(f"Train Shape: {X_train.shape}, Test Shape: {X_test.shape}")

# Feature Scaling for XGBoost
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ARIMA Model
arima_model = ARIMA(y_train, order=(5,1,0))
arima_fit = arima_model.fit()
arima_pred = arima_fit.forecast(steps=len(y_test))

arima_mse = mean_squared_error(y_test, arima_pred)

plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, arima_pred, label='ARIMA Forecast')
plt.title('ARIMA Forecast vs Actuals')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.legend()
plt.show()

print(f'ARIMA Mean Squared Error: {arima_mse}')

# Exponential Smoothing Model
es_model = ExponentialSmoothing(y_train, trend='add', seasonal='add', seasonal_periods=12)
es_fit = es_model.fit()
es_pred = es_fit.forecast(steps=len(y_test))

es_mse = mean_squared_error(y_test, es_pred)

plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, es_pred, label='Exponential Smoothing Forecast')
plt.title('Exponential Smoothing Forecast vs Actuals')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.legend()
plt.show()

print(f'Exponential Smoothing Mean Squared Error: {es_mse}')

# XGBoost Model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)
xgb_model.fit(X_train_scaled, y_train)
xgb_pred = xgb_model.predict(X_test_scaled)

xgb_mse = mean_squared_error(y_test, xgb_pred)

plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, xgb_pred, label='XGBoost Forecast')
plt.title('XGBoost Forecast vs Actuals')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.legend()
plt.show()

print(f'XGBoost Mean Squared Error: {xgb_mse}')

# Create directories if they do not exist
model_dir = '../models'
report_dir = '../reports'
if not os.path.exists(model_dir):
    os.makedirs(model_dir)
if not os.path.exists(report_dir):
    os.makedirs(report_dir)

# Ensure predictions are Series with the correct index
arima_pred_series = pd.Series(arima_pred, index=y_test.index)
es_pred_series = pd.Series(es_pred, index=y_test.index)
xgb_pred_series = pd.Series(xgb_pred, index=y_test.index)

# Save Predicted Results
predictions = pd.DataFrame({
    'date': y_test.index,
    'actual_units_sold': y_test,
    'arima_pred': arima_pred_series,
    'es_pred': es_pred_series,
    'xgb_pred': xgb_pred_series
})

predictions.to_csv(os.path.join(report_dir, 'predicted_results.csv'), index=False)
print("Predicted Results Saved.")

# Save Model Artifacts
joblib.dump(arima_fit, os.path.join(model_dir, 'arima_model.pkl'))
joblib.dump(es_fit, os.path.join(model_dir, 'exponential_smoothing_model.pkl'))
joblib.dump(xgb_model, os.path.join(model_dir, 'xgb_model.pkl'))
print("Models Saved.")
```
